```python
import numpy as np
from scipy.interpolate import lagrange
import matplotlib.pyplot as plt
```

```python
# Data points
x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 2, 3, 5, 8])

# Calculate Lagrange polynomial
poly = lagrange(x, y)

# Generate points for plotting the curve
x_new = np.linspace(min(x) - 0.5, max(x)*2 + 0.5, 100)
y_new = poly(x_new)

# Plotting
plt.figure(figsize=(8, 6))
plt.plot(x_new, y_new, label='Lagrange Interpolation Polynomial', color='blue')
plt.scatter(x, y, color='red', zorder=5, label='Given Points')

# Highlight points
for i, txt in enumerate(y):
    plt.annotate(f'({x[i]}, {txt})', (x[i], y[i]), textcoords="offset points", xytext=(0,10), ha='center')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Lagrange Interpolation Polynomial')
plt.legend()
plt.grid(True)
plt.savefig('lagrange_interpolation.png')
```
