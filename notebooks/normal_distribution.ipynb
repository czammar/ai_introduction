```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

$$ f(x \mid \mu, \sigma^2) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^2} $$

* $x$: The value of the random variable.
* $\mu$: The mean (average), which determines the location of the peak.
* $\sigma$: The standard deviation, which determines the spread or "fatness" of the curve.
* $\sigma^2$: The variance.
* $e$: Euler's number (approximately 2.71828).


```python
def plot_normal_distributions(params):
    """
    params: List of tuples [(mean, std_dev, color, label), ...]
    """
    # Create an x-axis range (3 standard deviations from the widest mean)
    x = np.linspace(-10, 10, 1000)
    
    plt.figure(figsize=(10, 6))

    for mu, sigma, color, label in params:
        # Calculate the probability density function (PDF)
        y = norm.pdf(x, mu, sigma)
        plt.plot(x, y, color=color, label=label, linewidth=2)
        
    plt.title('Normal Distributions with Different Parameters', fontsize=14)
    plt.xlabel('Value', fontsize=12)
    plt.ylabel('Probability Density', fontsize=12)
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()
```

```python
# Define sets of (mean, standard_deviation, color, label)
distributions = [
    (0, 1, 'blue', 'Standard (=0, =1)'),
    (0, 2, 'red', 'Wide (=0, =2)'),
    (2, 0.5, 'green', 'Narrow/Shifted (=2, =0.5)')
]

plot_normal_distributions(distributions)
```
